```
'''
Şmdi indirdiğimiz autoviz algoritmamızdan bir class oluşuralım.
'''
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
```

```
filename = "churn_data.csv"
sep = ","
dft = AV.AutoViz(
    filename, # Görselleştirmek istediğimiz dosya
    sep=",", # Dosyamız csv formatında olduğu için ","(virgül) koymalıyız. (csv'nin Türkçe açılımı virgülle ayrılan değerler demektir).
    depVar="churn", # Bağımlı(hedef) değişkenimizi buraya yazıyoruz.
    dfte=None, # bu, grafikleri çizmek için bir pandas dataframe yüklemek istemeniz durumunda kullanılır. Bu durumda, dosya adını boş bir dize olarak bırakın
    header=0, # dosyanızdaki başlık satırının satır numarası. Bu ilk satırsa, bu sıfır olmalıdır
    verbose=1,# Verbose = 0 ve Verbose = 1 görselleştirmek için, Verbose = 2 görselleri kayıt etmek için
    lowess=False, # Küçük veri kümelerinde kullanılır. Bunu büyük veri kümeleri için kullanmayın (yani 100.000'den fazla satır)
    chart_format="svg", # bu SVG, PNG veya JPG olabilir. Verbose = 2 seçeneğini kullandıysanız, bu formatta oluşturulan ve kaydedilen grafikleri alacaksınız. Grafikler oluşturmak ve daha sonra kullanmak için çok kullanışlıdır.
    max_rows_analyzed=27000, # Görselleştirme yapılacak maximum satır sayısı
    max_cols_analyzed=31, # Görselleştirme yapılacak maximum kolon sayısı
)
```
